# Step 2: Import Required Libraries

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
```

## Load model and set up filter


```python
model_name = "gpt2"   # You can also try "gpt2-medium" if GPU available

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

# Step 4: Filtering Mechanism
## Simple keyword-based filter for Python-related queries

```python
python_keywords = ["python", "code", "function", "class", "import", "def", "print", "list", "dict"]

def is_python_related(prompt: str) -> bool:
    prompt_lower = prompt.lower()
    return any(keyword in prompt_lower for keyword in python_keywords)
```

# Step 5: Response Generator

```python

def generate_response(prompt: str, max_length: int = 100):
    if not is_python_related(prompt):
        return "⚠️ I can only answer Python coding-related questions."

    inputs = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(inputs, max_length=max_length, do_sample=True, top_p=0.9, temperature=0.7)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
```

# Step 6: Testing

```python


test_prompts = [
    "How to write a Python function to reverse a string?",
    "What is the capital of France?",
    "Write Python code to generate Fibonacci numbers.",
    "Tell me a joke.",
    "How to use classes in Python?"
]

for prompt in test_prompts:
    print(f"📝 Prompt: {prompt}")
    print(f"🤖 Response: {generate_response(prompt)}\n{'-'*60}\n")
```
